````python
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole 
from rdkit.Chem import Draw
````

````python
molecule = Chem.MolFromSmiles('C[C@@H](CC1=CC=CC=C1)NC')
fig = Draw.MolToMPL(molecule, size=(100,100))
````

````python
molecule = Chem.MolFromSmiles('CCCCCC1=CC(=C2[C@H]3C=C(CC[C@@H]3C(OC2=C1)(C)C)C)O')
fig = Draw.MolToMPL(molecule, size=(200,200))
````
