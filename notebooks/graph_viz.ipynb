````python
import numpy as np
import nxviz as nv
import matplotlib.pyplot as plt
import networkx as nx

# Create a random Graph
G = nx.erdos_renyi_graph(30, 0.3)
G = nx.karate_club_graph()

# Create graph by hand
nodelist = [("a"), ("b"), ("c"), ("d"), ("e"), ("f")]
edgelist1 = [
    ("a", "b", {"weight": 0.1}),
    ("a", "c", {"weight": 0.2}),
    ("b", "d", {"weight": 0.6}),
    ("c", "d", {"weight": 0.7}),
    ("e", "d", {"weight": 0.8}),
    ("e", "f", {"weight": 1.0}),
]
edgelist2 = [
    ("a", "b", {"class": 1}),
    ("a", "c", {"class": 3}),
    ("b", "d", {"class": 8}),
    ("c", "d", {"class": 10}),
    ("e", "d", {"class": 5}),
    ("e", "f", {"class": 5}),
]

G = nx.Graph()
G.add_nodes_from(nodelist)
G.add_edges_from(edgelist1)
````

#### Common Graph with Nodes
We can see that this graph is unidirected.

````python
nx.draw(G);
plt.show();
````

#### Matrix Vizualization

````python
m = nv.plots.MatrixPlot(G)
plt.title('Matrix Plot')
m.draw()
plt.show()
````

#### ArcPlot Vizualization

````python
a = nv.plots.ArcPlot(G)
a.draw()
plt.title('Arc Plot')
plt.show()
````

#### Circos Plot

````python
#c = nv.plots.CircosPlot(G, edge_color="weight")
c = nv.plots.CircosPlot(G)
c.draw()
c.draw()
plt.show()
````
