#### Categorical Distribution

```python
def sample_categorical(batch_size, n_classes=10):
    # Get random values from 0 to n_classes
    cat = np.random.randint(0, n_classes, batch_size)
    # Convert to one-hot encoding
    cat_one_hot = np.eye(n_classes)[cat].astype('float32')
    # Convert to pytorch tensor
    #cat = torch.from_numpy(cat)
    return cat, cat_one_hot

samples, samples_one_hot = sample_categorical(5,10)
```
